## Step 1: Install Dependencies ⚙️

```python
# Step 1: Install JDK and setup Android SDK
# This block is safe to run multiple times.

# Install Java Development Kit
!apt-get install openjdk-17-jdk -y

# Define the Android SDK path
sdk_path = '/content/android_sdk'

# Only download and set up if the SDK path doesn't already exist
import os
if not os.path.exists(sdk_path):
    print("Setting up Android SDK...")
    # Download and unzip the Android command-line tools
    !wget https://dl.google.com/android/repository/commandlinetools-linux-11076708_latest.zip -O cmd-tools.zip
    !unzip -q cmd-tools.zip -d /content/temp_sdk

    # The tools are unzipped into a 'cmdline-tools' folder.
    # We will create the final SDK structure and move them into a 'latest' subdirectory.
    !mkdir -p {sdk_path}/cmdline-tools
    !mv /content/temp_sdk/cmdline-tools {sdk_path}/cmdline-tools/latest

    # Clean up temporary files
    !rm cmd-tools.zip
    !rm -rf /content/temp_sdk
    print("Android SDK setup complete.")
else:
    print("Android SDK is already set up.")
```

## Step 2: Set Up Environment Variables

```python
import os

# Set environment variables
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"
os.environ["ANDROID_SDK_ROOT"] = "/content/android_sdk"
os.environ["PATH"] += ":/content/android_sdk/cmdline-tools/latest/bin"
```
